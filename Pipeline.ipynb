```
f = pd.read_csv("drive/My Drive/ProcessedData.csv")
```

```
from sklearn.preprocessing import LabelEncoder
feature_cols = ['DepDelayMinutes','CRSArrTime','ArrDel15','CRSDepTime','DepDel15','windspeedKmph_y','winddirDegree_y', 'weatherCode_y',	'precipMM_y',	'visibility_y',	'pressure_y',	'cloudcover_y',	'DewPointF_y',	'WindGustKmph_y',	'tempF_y',	'WindChillF_y',	'humidity_y', 'airport_y']
X = f[feature_cols].copy(deep=True)
y = f['ArrDelayMinutes'].copy(deep=True)
le = LabelEncoder()
X['airport_y']=le.fit_transform(X['airport_y'])
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```
temp = pd.DataFrame(y_arrdel15_predict)
print(len(temp))
```

```
X_test_load = X_test_load.reset_index(drop = True)
X_test_load['ArrDel15'] = temp[0]
X_test_load
```

```
X_test_load['ArrDelayMinutes'] = pd.DataFrame(y_test)['ArrDelayMinutes'].reset_index(drop = True)
X_test_load
```

```
f_df = X_test_load.loc[X_test_load['ArrDel15'] == 1]
f_df
```

```
feature_cols = ['DepDelayMinutes','CRSArrTime','ArrDel15','CRSDepTime','DepDel15','windspeedKmph_y','winddirDegree_y', 'weatherCode_y',	'precipMM_y',	'visibility_y',	'pressure_y',	'cloudcover_y',	'DewPointF_y',	'WindGustKmph_y',	'tempF_y',	'WindChillF_y',	'humidity_y', 'airport_y']
X_test_final = f_df[feature_cols].copy(deep=True)
y_test_final = pd.DataFrame(f_df['ArrDelayMinutes'].copy(deep=True))
```

Random Forest

```
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor() 
regressor.fit(X_train, y_train)
y_pred_reg = regressor.predict(X_test_final)
```

```
import sklearn.metrics as metrics
import math
mae = metrics.mean_absolute_error(y_test_final, y_pred_reg)
mse = metrics.mean_squared_error(y_test_final, y_pred_reg)
rmse = math.sqrt(mse) 
r2 = metrics.r2_score(y_test_final,y_pred_reg)
print("MAE : ",mae)
print("MSE : ",mse)
print("RMSE : ",rmse)
print("R2 : ",r2)
```
