```
import json 
import pandas as pd
import os
import csv
```

```
f = pd.read_csv("drive/My Drive/ProcessedData.csv")
```

```
from sklearn.preprocessing import LabelEncoder
feature_cols = ['DepDelayMinutes','CRSArrTime','CRSDepTime','DepDel15','windspeedKmph_y','winddirDegree_y', 'weatherCode_y',	'precipMM_y',	'visibility_y',	'pressure_y',	'cloudcover_y',	'DewPointF_y',	'WindGustKmph_y',	'tempF_y',	'WindChillF_y',	'humidity_y', 'airport_y']
X = f[feature_cols].copy(deep=True)
y = f['ArrDel15'].copy(deep=True)
le = LabelEncoder()
X['airport_y']=le.fit_transform(X['airport_y']) 
```

**OVERSAMPLING**

SMOTE

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X_train, y_train = oversample.fit_resample(X_train, y_train)
```

```
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 10 , learning_rate = 0.05)
xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test.values)
y_pred_xgb = [round(value) for value in y_pred]
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_xgb))
```

ADASYN

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```
from imblearn.over_sampling import ADASYN

X_train, y_train = ADASYN(sampling_strategy = 'minority').fit_resample(X_train, y_train)
```

```
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 10 , learning_rate = 0.05)
xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test.values)
y_pred_xgb = [round(value) for value in y_pred]
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_xgb))
```

RANDOM OVER SAMPLING

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```
from imblearn.over_sampling import RandomOverSampler
oversample = RandomOverSampler(sampling_strategy='minority')
X_train, y_train = oversample.fit_resample(X_train, y_train)
```

```
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 10 , learning_rate = 0.05)
xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test.values)
y_pred_xgb = [round(value) for value in y_pred]
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_xgb))
```

**UNDERSAMPLING**

RANDOM UNDER SAMPLING

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(sampling_strategy='majority')
X_train, y_train = undersample.fit_resample(X_train, y_train)
```

```
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 10 , learning_rate = 0.05)
xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test.values)
y_pred_xgb = [round(value) for value in y_pred]
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_xgb))
```

NEAR MISS

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```
from imblearn.under_sampling import NearMiss
undersample = NearMiss(version=1, n_neighbors=3)
X_train, y_train = undersample.fit_resample(X_train, y_train)
```

```
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 10 , learning_rate = 0.05)
xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test.values)
y_pred_xgb = [round(value) for value in y_pred]
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_xgb))
```
