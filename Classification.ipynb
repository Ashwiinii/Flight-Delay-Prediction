```
import json 
import pandas as pd
import os
import csv
```

```
f = pd.read_csv("drive/My Drive/ProcessedData.csv")
```

```
from sklearn.preprocessing import LabelEncoder
feature_cols = ['DepDelayMinutes','CRSArrTime','CRSDepTime','DepDel15','windspeedKmph_y','winddirDegree_y', 'weatherCode_y',	'precipMM_y',	'visibility_y',	'pressure_y',	'cloudcover_y',	'DewPointF_y',	'WindGustKmph_y',	'tempF_y',	'WindChillF_y',	'humidity_y', 'airport_y']
X = f[feature_cols].copy(deep=True)
y = f['ArrDel15'].copy(deep=True)
le = LabelEncoder()
X['airport_y']=le.fit_transform(X['airport_y']) 
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```
X_test.to_csv('drive/My Drive/X_test.csv')
y_test.to_csv('drive/My Drive/y_test.csv')
```

Logistic

```
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter = 200, solver = 'saga')

logreg.fit(X_train,y_train)

y_pred_log=logreg.predict(X_test)
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_log))
```

Decision Tree

```
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X_train, y_train)
y_pred_dtree=dtree.predict(X_test)
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_dtree))
```

Extra Tree

```
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model = model.fit(X_train, y_train)
y_pred_extra = model.predict(X_test)
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_extra))
```

Random Forest

```
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, random_state = 0)
classifier.fit(X_train, y_train)
y_pred_random = classifier.predict(X_test)
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_random))
```

XGBoost

```
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 10 , learning_rate = 0.05)
xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test)
y_pred_xgb = [round(value) for value in y_pred]
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_xgb))
```
